```python
#import csv
df_init = pd.read_csv("D:\\simplon\\ia\\Projet de groupe octobre 2022\\191222\\librairie.csv")  
```

```python
#on enleve les clients pro 
df_nopro = df_init[(df_init["client_id"] !="c_1609") & (df_init["client_id"] !="c_4958") & (df_init["client_id"] !="c_6714")& (df_init["client_id"] !="c_3454")   ]
```

```python
#Potite heatmap pour visualiser les corrélation entre nos feature 
plt.figure(figsize=(4,4))
sns.heatmap(df_nopro[["age","total_spend","average_spend","ordered"]].corr(),annot=True)
```

On peu voir une corrélation entre age et average_spend et entre total_spend et ordered (ce qui est logique ) 

```python
#Hist pour de l'analyse exploratoire et tres visuel sur loi normale ou non 
df_nopro.hist()
```

```python
 #Boite moustache sex/total_spend , aucune différence notable a par les outliers 
sns.boxplot(data=df_nopro, x = "sex", y = "total_spend")
```

```python
#bins pour l'age
df_nopro['age_seg'] = pd.cut(df_nopro['age'], bins=[18,30,50,93], labels=['18-30','30-50','50-93'])
```

```python
#Model de reg pour obtenir les résidu 
mdl = statsmodels.formula.api.ols('average_spend ~ age_seg', data = df_nopro)
res = mdl.fit()
```

```python
#l'anova qu'on regardera apres avoir check nos 3 conditions mais qu'on fait la comme ça elle est a 
#La suite du model de rég
table = statsmodels.api.stats.anova_lm(res)
```

```python
# anova
table
```

```python
pplot = statsmodels.graphics.gofplots.ProbPlot(res.resid, fit = True)
pplot.qqplot(line = 's')
```

## Test de Shapiro-Wilk pour voir si notre echantillon suit une loi normal , et la ce n'est pas le cas 

```python
stats.shapiro(res.resid)
```

## On teste l'homosedasticité avec le test de levene , il n'est pas concluant 

```python
#stats.levene renvoie deux variable (stat et pvalue) donc on met deux variable
stat,p = stats.levene(df_nopro["average_spend"][df_nopro["age_seg"]== '18-30'],
             df_nopro["average_spend"][df_nopro["age_seg"]== '30-50'],
             df_nopro["average_spend"][df_nopro["age_seg"]== '50-93'])
```

```python
#Nous on cherche a regarder si la p value > alpha 
p
```

# KHI DEUX

```python
#Changement de type pour enlever l'ambiguité dans le type de variable ( qualitative)
df_nopro["fav"].astype('object')
```

```python
#On choisi no variable et on fait la tabe
X = "sex"
Y = "fav"
cont = df_nopro[[X, Y]].pivot_table(index=X, columns=Y, aggfunc=len).fillna(0).copy().astype(int)
```

```python
#Notre contingency table , on peu voir a l'oeil que globalement les ordre de grandeur sont respecter 
cont
```

```python
#On fait le chi2 sur notre table et on recupere les 3 variable que renvoie stats.chi2_contingency 
st_chi2, st_p, st_dof, st_exp = stats.chi2_contingency(cont)
```

```python
# Statistique khi2
print("Stat khi2 :",st_chi2 )
 
# Nombe de degrés de liberté

print("Nombre de degrés de liberté :",st_dof  )
# Pvalue : l'unique valeur nécessaire pour prendre notre décision

print("Pvalue :",st_p )
```

Conclusion :les variables sont indépendantes,
